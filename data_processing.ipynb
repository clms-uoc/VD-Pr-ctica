## Datasets saved for web

```python
import os
```

```python
def prepare_color_scaling(df, df_name):
    df = df[['crag', 'sector', 'name', 'grade_fra', 'Country']]
    df.to_csv(f'datasets/flourish/{df_name}_atrributes.csv')
    summary = df.groupby(['Country', 'grade_fra']).size().reset_index(name='route_count')
    summary['color_intensity'] = summary['route_count']  

    summary.to_csv(f'datasets/flourish/{df_name}_color_intensity.csv', index=False)
```

```python
root_path = 'datasets/continents/'
for df_path in os.listdir(root_path):
    df = pd.read_csv(root_path+df_path)
    df_name = os.path.basename(df_path).split('.')[0]
    prepare_color_scaling(df, df_name)
```

## Climber df

```python
import pandas as pd
```

```python
climber_df = pd.read_csv("datasets/climber_df.csv")
countries_df = pd.read_csv("datasets/countries_codes_and_coordinates.csv")
grades_df = pd.read_csv("datasets/grades_conversion_table.csv")

alpha3_to_country_map = {
    'SWE': 'Sweden',
    'NOR': 'Norway',
    'BEL': 'Belgium',
    'CAN': 'Canada',
    'GBR': 'United Kingdom',
    'NLD': 'Netherlands',
    'USA': 'United States',
    'DEU': 'Germany',
    'ITA': 'Italy',
    'FRA': 'France',
    'POL': 'Poland',
    'ZAF': 'South Africa',
    'PRT': 'Portugal',
    'AUS': 'Australia',
    'CZE': 'Czech Republic',
    'AUT': 'Austria',
    'ESP': 'Spain',
    'HRV': 'Croatia',
    'MEX': 'Mexico',
    'BRA': 'Brazil',
    'CHE': 'Switzerland',
    'SVN': 'Slovenia',
    'DNK': 'Denmark',
    'RUS': 'Russia',
    'FIN': 'Finland',
    'other': 'Other'  
}

climber_df["country"] = climber_df["country"].map(alpha3_to_country_map).fillna("Other")
def map_grades(row, grades_df):
    rounded_first = round(row["grades_first"])
    rounded_last = round(row["grades_last"])
    rounded_mean = round(row["grades_mean"])
    return (
        grades_df.loc[grades_df["grade_id"] == rounded_first, "grade_fra"].values[0],
        grades_df.loc[grades_df["grade_id"] == rounded_last, "grade_fra"].values[0],
        grades_df.loc[grades_df["grade_id"] == rounded_mean, "grade_fra"].values[0],
    )

climber_df[["grade_first_fra", "grade_last_fra", "grade_mean_fra"]] = climber_df.apply(
    lambda row: pd.Series(map_grades(row, grades_df)), axis=1
)

climber_df.drop(columns=["grades_first", "grades_last", "grades_mean"], inplace=True)

climber_df.sort_values(by="country", inplace=True)

climber_df.to_csv("processed_climber_df.csv", index=False)

print(climber_df)
```
